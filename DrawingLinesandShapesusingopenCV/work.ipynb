```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
#Create a empty canvas

height , width  = 500 , 700

blue = (255 , 127 , 0 )

canvas  = np.full((height , width ,  3 ) , blue , dtype=np.uint8)
canvas_rgb = cv2.cvtColor(canvas , cv2.COLOR_BGR2RGB)
plt.imshow(canvas_rgb)
```

```python
plt.imshow(canvas)
```

```python
image_dag = cv2.imread("./mountain.jpg")
plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.imshow(image_dag)
plt.title("BGR")
plt.axis("off")
plt.subplot(2,2,2)
image_dag_rgb = cv2.cvtColor(image_dag , cv2.COLOR_BGR2RGB)
plt.imshow(image_dag_rgb)
plt.title("RGB")
plt.axis("off")
```

# DRAW A LINE 


```python
plt.imshow(canvas_rgb)
```

```python
cv2.line(canvas_rgb , (50,50)  , (650 , 450),(255,255,255),5)
plt.imshow(canvas_rgb)
```

# Draw Rectangle 

```python
cv2.rectangle(canvas_rgb , (10 ,10)  , (200,200)  , (0,255,255) , -1 )
plt.imshow(canvas_rgb)
```

# Draw Circle


```python
cv2.circle(canvas_rgb , (100 , 300) , 50 , (255 , 0, 0) , 10 )
plt.imshow(canvas_rgb)
```

# Draw an Ellips 

```python
cv2.ellipse(canvas_rgb , (350 ,350) , (100 ,50 ) , 45 , 0 ,360 , (0,255,0) , -1)
plt.imshow(canvas_rgb)
```

# Draw Polygon



```python


points = np.array([[150,400] , [300,400] , [225 , 300]] , np.int32) 
new_reshaped_point = points.reshape((-1,1,2))
print(new_reshaped_point)
```

```python
cv2.polylines(canvas_rgb , [new_reshaped_point] , True  , (255,  0,  0 )  , 3 )
plt.imshow(canvas_rgb )
```
