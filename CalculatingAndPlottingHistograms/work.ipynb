```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
#RGB 
image = cv2.imread("./giraffe.jpg")
image_rgb = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
channels = cv2.split(image_rgb)
colors = ["red" , "green" ,"blue"   ]
for channel , color  in zip(channels , colors):
    hist = cv2.calcHist([channel] , [0] , None ,  [256] , [0,256] )
    plt.plot(hist , color=color)

plt.title("color histogram")
plt.xlabel("pixel intentisity")
plt.ylabel("frequency")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# HSV'ye evir
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(image_hsv)

# Histogramlar
h_hist = cv2.calcHist([h], [0], None, [100], [0, 180])   # Hue 0-179 arasdr
s_hist = cv2.calcHist([s], [0], None, [256], [0, 256])
v_hist = cv2.calcHist([v], [0], None, [256], [0, 256])

# izim
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Hue Histogram
axes[0].plot(h_hist, color="orange")
axes[0].set_title("Hue Histogram")
axes[0].set_xlabel("Hue Value")
axes[0].set_ylabel("Frequency")

# Saturation Histogram
axes[1].plot(s_hist, color="green")
axes[1].set_title("Saturation Histogram")
axes[1].set_xlabel("Saturation")
axes[1].set_ylabel("Frequency")

# Value Histogram
axes[2].plot(v_hist, color="gray")
axes[2].set_title("Value Histogram")
axes[2].set_xlabel("Brightness (Value)")
axes[2].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
