```python
import cv2
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread("./giraffe.jpg")
image_rgb = cv2.cvtColor(image  , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

**FLIP**


```python
#1 Horizantally
#0 vertically
# -1 both axes
flips =[]
flips.append((cv2.flip(image_rgb ,1) , "HORIZANTAL"))
flips.append((cv2.flip(image_rgb ,0) ,  "vertical"))
flips.append((cv2.flip(image_rgb ,-1), "both"))
plt.figure(figsize=(8,8))
x = 2
plt.subplot(2,2,1)
plt.imshow(image_rgb)
plt.title("original")
for i   in flips:
    plt.subplot(2,2,x)
    plt.imshow(i[0]) 
    plt.title(i[1])
    plt.axis("off")
    x = x+1

plt.tight_layout()
plt.show()
```

# Rotate 


```python
rotate_90_clockwise = cv2.rotate (image_rgb  ,cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate (image_rgb  ,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180_clockwise = cv2.rotate (image_rgb  ,cv2.ROTATE_180)
plt.figure(figsize=(8,8))
rotation = [rotate_90_clockwise, rotate_180_clockwise , rotate_90_counterclockwise]
names = ["rotate_90_clockwise", "rotate_180_clockwise" , "rotate_90_counterclockwise"]
x= 2
plt.subplot(2,2,1)
plt.imshow(image_rgb)
plt.title("original")
for i , j  in zip(rotation , names):
    plt.subplot(2,2,x)
    x = x+1 
    
    plt.imshow(i)
    plt.title(j)
    plt.axis("off")
plt.tight_layout()
plt.show()
```
