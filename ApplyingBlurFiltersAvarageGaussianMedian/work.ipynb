Bluring    reduce noise   ,smoothing images
mesela bir zigzag çizgiyi düz bir çizgiye benzetebilir 
blurring için farklı yöntemler var avarage blur meesela  o zigzagı düzelten gibi düşün 
avarageda etrafın ortalaması alınıyor 
gaussianda centera gittikçe weighted avaragelar artıyor  
median blurda da medianı alıp koyuyo  gaussianda kernel  tek sayı olmak zorunda nedenini de açıkla

```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
image = cv2.imread("./giraffe.jpg")
image_rgb = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
image_grey = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.imshow(image_grey , cmap="gray")
```

```python
kernel_s = 11  


average_blur = cv2.blur( image_grey ,  (kernel_s ,  kernel_s))

plt.imshow(average_blur , cmap="gray")
```

```python
gauss_blur = cv2.GaussianBlur(image_grey , ( kernel_s , kernel_s) ,10 , 0)    # sigma x ve sigmay üzerine oluyo 
plt.imshow(gauss_blur ,   cmap="gray")
```

```python
median_blur = cv2.medianBlur(image_grey , 31 )  # 3 is kernel size  so it must be a square 
plt.imshow(median_blur , cmap="gray")
```

```python
plt.figure(figsize=(12,8) )
plt.subplot(2,2,1)
plt.title("original")
plt.imshow(image_rgb , cmap="gray")
plt.axis("off")
plt.subplot(2,2,2)
plt.title("average")
plt.imshow(average_blur , cmap="gray")
plt.axis("off")
plt.subplot(2,2,3)
plt.title("gaussian")
plt.imshow(gauss_blur , cmap="gray")
plt.axis("off")
plt.subplot(2,2,4)
plt.title("")
plt.imshow(median_blur , cmap="gray")
plt.axis("off")
plt.tight_layout()
plt.show()
```
