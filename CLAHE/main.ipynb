```python
import cv2 
import matplotlib.pyplot as plt
```

```python
image_bgr = cv2.imread("./giraffe.jpg")
plt.imshow(image_bgr)
```

```python
image_gray = cv2.cvtColor( image_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray , cmap="gray")
```

```python
equalised_image = cv2.equalizeHist(image_gray)
plt.imshow(equalised_image , cmap="gray")
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0  ,tileGridSize=(4,4))
image_try1 = clahe.apply(image_gray)
plt.imshow(image_try1 , cmap="gray")
```

```python
image_yuv = cv2.cvtColor(image_bgr , cv2.COLOR_BGR2YUV)
plt.imshow(image_yuv)
```

```python
image_yuv[:,:,0] = clahe.apply(image_yuv[:,:,0])
temp = cv2.cvtColor(image_yuv  , cv2.COLOR_YUV2BGR)
image_rgb_x = cv2.cvtColor(temp , cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb_x)
```

```python
image_rgb = cv2.cvtColor(image_bgr , cv2.COLOR_RGB2BGR)
plt.imshow(image_rgb)
```
